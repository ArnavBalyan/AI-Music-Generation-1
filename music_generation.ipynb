```python
from music21 import instrument,chord,note,stream,converter
import numpy as np
import glob
import pickle
from keras.utils import np_utils
```

```python
#PREPROCESS MUSIC FILES
notes=[]

for file in glob.glob("midi_songs/*.mid"):
  midi=converter.parse(file)
  print("PARSING %s"%file)
  elements_to_parse=midi.flat.notes

  for e in elements_to_parse:
    if isinstance(e,note.Note):
      notes.append(str(e.pitch))
    elif isinstance(e,chord.Chord):
      notes.append("+".join(str(c) for c in e.normalOrder))  
```

```python
len(notes)
```

```python
with open ("notes_generated",'wb') as f:
  pickle.dump(notes,f)
```

```python
n_vocab=len(set(notes)) #unique notes
```

```python
pitchnames=sorted(set(notes))
```

```python
ele_to_int=dict((ele,num) for num,ele in enumerate(pitchnames))
print(ele_to_int)
```

```python
#Preparing sequential data
network_input=[]
network_output=[]
seq_length=100
for i in range(len(notes)-seq_length):
  seq_in=notes[i:i+seq_length]
  seq_out=notes[i+seq_length]
  network_input.append([ele_to_int[c] for c in seq_in])
  network_output.append(ele_to_int[seq_out])
```

```python
len(network_input)
```

```python
network_input=np.reshape(network_input,(len(network_input),seq_length,1))
print(network_input.shape)
```

```python
normalized_network_input=network_input/float(n_vocab)
```

```python
network_output=np_utils.to_categorical(network_output) # one hot encoding for output vector to be fed in LSTM model
```

```python
print(network_output.shape)
```

```python
#MODEL ARCHITECTURE
from keras.models import Sequential,load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
model=Sequential()
model.add(LSTM(512,input_shape=(normalized_network_input.shape[1],normalized_network_input.shape[2]),return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=False))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam")
  
```

```python
#I TRAIN THIS MODEL USING GOOGLE COLAB AND SAVE IT BEST MODEL IN FILE SO THAT OUR MODEL NEED NOT TO BE TRAINED AGAIN
checkpoint=ModelCheckpoint("new_weights.hdf5",mode='min',save_best_only=True,monitor='loss',verbose=0)
hist=model.fit(normalized_network_input,network_output,epochs=100,callbacks=[checkpoint],batch_size=64)
```

```python
model=load_model("new_weights.hdf5")
```

```python
#PREDICTION
seq_length=100
network_input=[]
for i in range(len(notes)-seq_length):
  seq_input=notes[i:i+seq_length]
  network_input.append([ele_to_int[ch] for ch in seq_input])  
```

```python
start=np.random.randint(len(notes)-1)
start
```

```python
#int to ele mapping
int_to_ele=dict((num,ele)for num,ele in enumerate(pitchnames)) 
```

```python
pattern=network_input[start]
prediction_output=[]
for i in range(200):
  prediction_input=np.reshape(pattern,(1,seq_length,1))
  prediction_input=prediction_input/float(n_vocab)
  prediction=model.predict(prediction_input,verbose=0)
  max_idx=np.argmax(prediction)
  result=int_to_ele[max_idx]
  prediction_output.append(result)
  pattern.append(max_idx)
  pattern=pattern[1:]
  
```

```python
print(len(prediction_output))
```

```python
prediction_output
```

```python
#CREATING MIDI FILE
offset=0
output_notes=[]
for pattern in prediction_output:
  #if pattern is chord
  if ('+' in pattern) or pattern.isdigit():
    notes_in_chord=pattern.split('+')
    temp_notes=[]
    for curr in notes_in_chord:
      new_note=note.Note(int(curr))
      new_note.storedInstrument=instrument.Piano()  
      temp_notes.append(new_note)

    new_chord=chord.Chord(temp_notes)
    new_chord.offset=offset
    output_notes.append(new_chord)
  else:
     #if pattern is note
     new_note=note.Note(pattern)
     new_note.offset=offset
     new_note.storedInstrument=instrument.Piano()  
     output_notes.append(new_note)

  offset+=0.5  
```

```python
#CREATE A STREAM OBJECT
midi_stream=stream.Stream(output_notes)
midi_stream.write('midi',fp="output_test_aniket.mid")
```
